# **Understand Django Project Folder Structure**

### **1. Project Root Directory (myproject/)**

This is the top-level folder, named after the project (in this case, myproject/). It contains all the code required for the project to run. Here's what you will see inside:

```bash
map
myproject/
â”‚
â”œâ”€â”€ manage.py
â””â”€â”€ myproject/
    â”œâ”€â”€ __init__.py
    â”œâ”€â”€ settings.py
    â”œâ”€â”€ urls.py
    â”œâ”€â”€ asgi.py
    â””â”€â”€ wsgi.py
```

### **2. Key Components in the Project Structure**

#### A. **`manage.py`**

- This is a **command-line utility** that allows you to interact with the Django project.

- You use this file to run commands like:

- Start the development server:

  ```bash
  python manage.py runserver

  ```

- Make migrations:
  ```bash
  python manage.py makemigrations
  ```
- Apply migrations:
  ```bash
  python manage.py migrate
  ```

#### **B. myproject/(Inner Project Directory)**

This directory has the same name as the outer project folder, but it contains the core settings and configurations for your Django project.

## **3. Files in the Inner Project Directory**

#### **`__init__.py`**

- An empty file that indicates this directory is a Python package.
- This allows Python to treat this directory as a package, making imports work smoothly.

#### **`settings.py`**

- Contains **all configurations** for the project, including:
  - Installed apps
  - Middleware settings
  - Database configurations
  - Static and media file paths
- This is where you register new apps and modify project-wide settings.

```bash
from django.urls import path
from myapp import views

urlpatterns = [
    path('', views.home, name='home'),
]
```

#### **`asgi.py`**

- Stands for **Asynchronous Server Gateway Interface**.
- Used when deploying the project with ASGI servers for handling asynchronous web applications.

#### **`wsgi.py`**

- Stands for **Web Server Gateway Interface**.
- This file helps with the deployment of traditional synchronous web applications using servers like Gunicorn or Apache.

### **4. App Directory (Optional: If You Create an App)**

If you create an app within the project using:

```bash
$ python manage.py startapp myapp
```

You will get an additional folder like this:

```bash
map
myapp/
â”‚
â”œâ”€â”€ __init__.py
â”œâ”€â”€ admin.py
â”œâ”€â”€ apps.py
â”œâ”€â”€ models.py
â”œâ”€â”€ tests.py
â””â”€â”€ views.py
```

#### **Hereâ€™s what each file does:**

- **`admin.py`**: Configure how your models appear in the Django admin interface.

- **`apps.py`**: Defines configuration settings for the app.

- **`models.py`**: Define the data models (tables) here.

- **`tests.py`**: Write unit tests for your app here.
views.py: Contains the logic to handle requests and responses.

### **5. Migrations Folder**

When you create migrations for your models, Django generates files inside a migrations/ folder within the app. These files keep track of changes to your models and allow Django to apply them to the database.

# **Create a GitHub repo and push the initial setup.**

### **Step 1: Create a Repository on GitHub**

1. Log in to your GitHub account.
2. Click on the New button next to "Repositories" or go to this link.
3. Enter a Repository name (e.g., DJANGO_AND_GENAI).
4. Add an optional Description.
5. Choose Public or Private for the visibility.
6. Check Initialize this repository with a README (optional).
7. Add a .gitignore file (choose Python or relevant language if required).
8. Choose a License (e.g., MIT License).
9. Click Create repository.

### **Step 2: Clone the Repository to Your Local Machine**

Open your terminal (Command Prompt, PowerShell, or Terminal on macOS/Linux), and run:

```bash
git init
```
this command will initiate and creates a space for github repo.

```bash
git clone https://github.com/your-username/DJANGO_AND_GENAI.git
```
this will make a clone of the project that you created on your github into your local repo.

```bash
cd DJANGO_AND_GENAI
```
change dir into the created folder

### **Step 3: Add Initial Files**

Add any necessary files for your project, such as:

```bash
mkdir src
```
create an empty dir.


### **Step 4: Add, Commit, and Push Changes**

```bash
git add .
```
add all files(changes that you made) to git stage area

```bash
git commit -m "created an empty folder src"
```
commit the changes that you made and make ready to push to github repo

```bash
git push origin main
```
push the files into main branch

### **Step 6: Verify on GitHub**

Go to your GitHub repository page, and you should see the files pushed successfully.
